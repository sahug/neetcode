**Neetcode - 2 Pointers - Container With Most Water**

You are given an integer array **height** of length **n**. There are **n** vertical lines drawn such that the two endpoints of the **ith** line are **(i, 0)** and **(i, height[i])**.

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the *maximum amount of water a container can store.*

Notice that you may not slant the container.

**Example 1:**

image.png

```
Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. 
In this case, the max area of water (blue section) the container can contain is 49.
```

**Example 2:**
```
Input: height = [1,1]
Output: 1
```

````python
from typing import List
class Solution:

  def containerWithMostWater1(self, nums: List[int]) -> List[int]:
    start, end = 0, len(nums) - 1
    container = list()
    for i in range(len(nums)):
      for j in range(i+1, len(nums)):
        height = min(nums[i], nums[j])
        width = j - i
        area = height * width
        container.append(area)
    print(max(container))

  # 2 Pointer
  def containerWithMostWater2(self, height: List[int]) -> int:
    l, r = 0, len(height) - 1
    res = 0
    
    while l < r:
        res = max(res, min(height[l], height[r]) * (r - l))
        if height[l] < height[r]:
            l += 1
        elif height[r] <= height[l]:
            r -= 1
    print(res)
````

````python
solution = Solution()
solution.containerWithMostWater1([1,8,6,2,5,4,8,3,7])
solution.containerWithMostWater1([1,1])

solution.containerWithMostWater2([1,8,6,2,5,4,8,3,7])
solution.containerWithMostWater2([1,1])
````
