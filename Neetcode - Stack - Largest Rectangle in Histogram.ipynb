**Neetcode - Stack - Largest Rectangle in Histogram**


Given an array of integers **heights** representing the histogram's bar height where the width of each bar is **1**, *return the area of the largest rectangle in the histogram.*

**Example 1:**
image.png

```
Input: heights = [2,1,5,6,2,3]
Output: 10
Explanation: The above is a histogram where width of each bar is 1.
The largest rectangle is shown in the red area, which has an area = 10 units.
```

**Example 2:**
image.png

```
Input: heights = [2,4]
Output: 4
```

````python
from typing import List
class Solution():

  # Calculate the time each time will reach the target. (Target - Position/Speed)
  # If the cars behind has less time than car in from they will cloide and become one flee and so on.

  def largestRectangleArea(self, heights: List[int]) -> int:    
      maxArea = 0
      stack = [] # pair: (index, height)
      
      for i, h in enumerate(heights):
          start = i
          while stack and stack[-1][1] > h:
              index, height = stack.pop()
              maxArea = max(maxArea, height * (i - index))
              start = index
          stack.append((start, h))
      
      for i, h in stack:
          maxArea = max(maxArea, h * (len(heights) - i))
      return maxArea
````

````python
solution = Solution()
print(solution.largestRectangleArea([2,1,15,15,15,2,3,20,15,15]))
print(solution.largestRectangleArea([2,1,5,6,2,3]))
print(solution.largestRectangleArea([2,4]))
````
