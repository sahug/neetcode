**Neetcode - Binary Search - Search a 2D Matrix**


Write an efficient algorithm that searches for a value target in an m x n integer matrix matrix. This matrix has the following properties:

Integers in each row are sorted from left to right.
The first integer of each row is greater than the last integer of the previous row.

**Example 1:**

image.png

```
Input: matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 3
Output: true
```
**Example 2:**

image.png
```
Input: matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 13
Output: false
```

````python
from typing import List
class Solution():

  def search2DMatrix(self, matrix: List[List[int]], target) -> bool:    
    r_m, l_m = len(matrix) - 1, 0
    r_l, l_l = len(matrix[0]) - 1, 0
    
    while l_m <= r_m:
      mid_m = r_m + l_m //2      
      if target < matrix[mid_m][l_l]:
        r_m = mid_m - 1
      elif target > matrix[mid_m][r_l]:
        l_m = mid_m + 1
      else:
        while l_l <= r_l:        
          mid_l = r_l + l_l //2        
          if target == matrix[mid_m][mid_l]:
            return True
          elif target > matrix[mid_m][mid_l]:
            l_l = mid_l + 1          
          elif target < matrix[mid_m][mid_l]:
            r_l = mid_l - 1
    return False
````

````python
solution = Solution()
print(solution.search2DMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]], 3))
print(solution.search2DMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]], 13))
print(solution.search2DMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]], 60))
print(solution.search2DMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]], 1))
print(solution.search2DMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]], 12))
````
